Me interesa limpiar los datos de Age y Cabin
Necesito poner como valor Age: 0 y a Cabin: CNada y Embarked : J

```python
df_filtrado = df.fillna({"Age":0,"Cabin": 'CNada', "Embarked": 'J'})
df_filtrado.count()
df_filtrado.info()
```


podemos agrupar por GÃ©nero/Sex

```python
df_generos =df.groupby("Sex")
df_generos.describe()
```

Vamos a extraer los datos con informaciÃ³n en variables
Quiero primero, con un print, mostrar el nÃºmero de pasajeros totales

```python
pasajeros = df.shape[0] #nos da 2 valores, el primero es el nÂº de pasajeros por eso se pone 0 (filas y columnas)
print(f"NÃºmero de pasajeros: {pasajeros}")
```

vamos a sacar los supervivientes

```python
pasajeros= df.shape[0]
supervivientes = df[df.Survived == 1]
num_survived = supervivientes.shape[0]
fallecidos = df.query("Survived == 0")
num_fallecidos = fallecidos.shape[0]
print(f" Numero de pasajeros: {pasajeros}")
print(f" Numero de supervivientes: {num_survived}")
print(f" Numero de fallecidos: {num_fallecidos}")
```

Almacenar los datos que tenemos en un Diccionario

```python
data = {
    'Datos': ['pasajeros totales','Supervivientes', 'fallecidos']
         ,'Valores':[pasajeros,num_survived,num_fallecidos]
         }
df_estadistica= pd.DataFrame.from_dict(data)
df_estadistica
```

A continuaciÃ³n creamos un grÃ¡fico de Pie con el DataFrame de estadÃ­stica.
Para mostrar los datos agrupamos por la columna llamada Datos

```python
df_grafico= df_estadistica.groupby('Datos').sum()
df_grafico.plot(kind='pie', y='Valores')
```
