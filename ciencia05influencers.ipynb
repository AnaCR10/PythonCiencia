Dicho m√©todo nos permite realizar acciones sobre una columna (por ejemplo) fila a fila

def miFuncion(valores):
    resultado = valores * 5+2
    return resultado

df['COLUMNA'].apply(miFunction)

```python
def multiplicarDoble(menciones):
    doble = menciones * 2
    return doble

#agregamos una nueva columna al dataframe con el doble de menciones
df['doble_mentions']=df['mentions'].apply(multiplicarDoble)
df.head()
```

vamos a crear una nueva columna que sea la Popularidad del influencer.
La popularidad est√° expuesta por dividir los seguidores entre cuantas personas estoy siguiendo
(followees/followers)
Para ello hay que enviar los datos de toda la fila y luego recupero las posiciones de las columnas
Para poder enviar la fila del DataFrame en lugar de un valor de una columna se utiliza axis como 
par√°metro dentro de Apply
axis = 1 # significa enviar la fila

```python
def popularidad(fila):
    followers = fila['followers']
    followees = fila['followees']
    resultado = followees/followers
    return resultado

df['popularidad'] = df.apply(popularidad, axis=1)
df.head()
df[df['popularidad']>=15]
```

vamos a agrupar por el campo COUNTRY y vamos a mostrar la media de dicho grupo
Dentro de nuestro DataFream tenemos textos, tenemos que poner numeric_only para recuperar 
solo las columnas con datos num√©ricos

```python
df.groupby(['country']).mean(numeric_only=True)
```

Imaginemos que necesitamos recuperar la media y la suma de los campos
MEDIA
df.groupby(['country']).mean(numeric_only=True)
SUMA
df.groupby(['country']).sum(numeric_only=True)
Tenemos una funci√≥n para los grupos que nos permite mostrar m√°s de un resumen en la misma
instrucci√≥n:  agg({"max", "min"})
Tambi√©n nos permite incluir que columnas podemos mostrar dentro de agg
agg({ columna: "max", columna : "min"})

```python
df.groupby(['country']).agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max"
    })
```

podemos filtrar por un grupo creado, quiero mostrar los datos que tengan m√°s de 5000 seguidores


```python
grupo = df.groupby(['country']).agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max"
    })
grupo[grupo['followers']> 5000]
```

al igual que podemos leer datos podemos exportar datos.
Vamos a almacenar el grupo dentro de un csv

```python
grupo.to_csv('data/grupo.csv')
grupo.to_json('data/grupo.json')
grupo.to_html('data/grupo.html')
```
